```
import cv2
import cvlib
from cvlib.object_detection import draw_bbox
from google.colab.patches import cv2_imshow

```

Use Open CV to load and display the image.

```
img = cv2.imread("/content/sample_data/car.jpeg")
cv2_imshow(img)
```

Use detect_common_objects function to detect the object. This in build function returns bounding box or bbox and its corresponding labels. Use draw_bbox function to draw bbox and label on a given image. Then display the image. 

```
bbox,label,conf = cvlib.detect_common_objects(img)
output_img = draw_bbox(img,bbox,label,conf)
cv2_imshow(output_img)
```

We need to count number of bus,car, and motorcycle in the given image. First print or display label to verify number of different objects it contains. Use simple for loop and if loop to count number of occurances of a particular type of objects. It will give you number of object in a given input.

```
counter_bus= 0
counter_car = 0
counter_motorcycle = 0
for i in range(len(label)):
     if label[i] == "bus":
        counter_bus +=1
     if label[i] == "car":
        counter_car +=1
     if label[i] == "motorcycle":
        counter_motorcycle +=1      

print(f"Number of cars {counter_car}, number of buses {counter_bus}, and number of motorcycle {counter_motorcycle} ")
```
