###### Define Binomial distribution

The Bernoulli distribution is the simplest form of Binomial distribution. If number of trial n=1 then Binomial distribution becomes Bernoulli distribution. For instance, X refers the probability to get head for a fair coin toss. X ~ Bern(1/2)

```python
def binomial_dist(x):
    x= np.random.binomial(n=x,p=0.5,size=200)
    #print(x)
    sns.distplot(x,hist=True,kde=True)
    plt.title("Binomial Distribution")
    plt.show()
```

#### Bernoulli Distribution 

```python
binomial_dist(1)
```

##### Call binomial_dist() to plot Binomial distribution 

```python
binomial_dist(20)
```

##### Define Normal distribution 

If we increase sample size then it will follow normal distribution X ~ N($\mu,\sigma ^2$).The units for the normal distribution curve are denoted by Z and are called the Z-values or Z Scores. It is also known as standard units or standard scores. Z score or Z value represents x-axis. Z=1 means 1 standards deviation from the right of the mean. We can define Z as follows:

$$ Z = {X - \mu \over \sigma} $$

```python
def normal_dist():
    x= np.random.normal(loc=50,scale=5,size=200)
    #print(x)
    sns.distplot(x,hist=True,kde=True)
    plt.title("Normal Distribution")
    plt.show()
```

##### Call normal_dist() to plot Normal distribution 

```python
normal_dist()
```

#### Poisson distribution can be defined as follows:

$$ P(X)  =  {\lambda ^ X * e ^ -\lambda \over X!}$$

If n is the number of trial and p is the probability of success then $\lambda = mean = n*p$

If n ≥ 20 and p < 0.05 then Poisson distribution approaches to Binomial distribution. 

 
#### Define Poisson Distribution 

```python
def poisson_dist():
    x=np.random.poisson(lam=50.0,size=2000)
    sns.distplot(x,hist=True,kde=True)
    plt.title("Poisson Distribution")
    plt.show()
```

#### Call poisson_dist() to plot Poisson distribution 

```python
poisson_dist()
```

#### Define Unifrom distribution

Let us consider a kid is throwing a ball in the room. It is expected that ball can hit anywhere in the room. Ball is equally likely to hit anywhere. We can say that ball has uniform distribution i.e. Unif(a,b)

```python
def uniform_dist():
    x = np.random.uniform(low=0.0,high=25,size=2000)
    sns.distplot(x,hist=True,kde=True)
    plt.title("Unifrom Distribution")
    plt.show()
```

#### Call uniform_dist() to plot Uniform distribution 

```python
uniform_dist()
```

#### Define Exponential distribution 

Let us consider a continuous random variable x, with its p.d.f defined by
        
                    
                         

$$ f(x) = \mu * e ^ {- \mu * x} , if x > 0 $$ 
$$ f(x) = 0 , otherwise $$

where $\mu > 0$, is said to have a exponential distribution

```python
def exponential_dist():
    x=np.random.exponential(scale=1.0,size=20)
    sns.distplot(x,hist=True,kde=True)
    plt.title("Exponential Distribution")
    plt.show()
```

#### Call exponential_dist() to plot Exponential distribution 

```python
exponential_dist()
```

#### Define Gamma distribution 

It is also known as Erlang Distribution if it assumes only non-negative values and its p.d.f is given by 



$$ f(x) = {\mu * (\mu x)^ {n-1} * e ^ {- \mu x} \over (n-1)!} , if  0 < x < inf $$

$$ f(x) =  0, otherwise $$

```python
def gamma_dist():
    x=np.random.gamma(shape=1,scale=1.0,size=200)
    sns.distplot(x,hist=True,kde=True)
    plt.title("Gamma Distribution")
    plt.show()
```

#### Call gamma_dist() to plot Gamma distribution

```python
gamma_dist()
```

#### Define Beta distribution 

A random variable x is said to have beta distribution if its density function is given by

$$ f(x) = { x ^ {p-1} (1-x) ^ {q-1} \over \beta (p,q) }, if 0 < x < 1 ; p > 0 , q > 0 $$

$$ f(x) = 0, if x \leq 0  and  x \geq 1 $$

```python
def beta_dist(x,y):
    x1=np.random.beta(x,y,size=200)
    sns.distplot(x1,hist=True,kde=True,label= f"({x},{y})")
    plt.title("Beta Distribution")
    plt.legend()
    plt.show()
```

#### Call beta_dist() to plot Beta distribution 
#### Beta distribution (0.5,0.5)

```python
beta_dist(0.5,0.5)
```

#### Beta distribution (2,1) 

```python
beta_dist(2,1)
```

#### Beta distribution (2,8) 

```python
beta_dist(2,8)
```

#### Beta distribution (5,5) 

```python
beta_dist(5,5)
```

#### Comapre All Distributions 

```python
def compare():
    x= np.random.normal(loc=10,scale=1,size=1000)
    y= np.random.binomial(n=20,p=0.5,size=1000)
    z=np.random.poisson(lam=10.0,size=1000)
    uniform = np.random.uniform(low=0.0,high=25,size=1000)
    sns.distplot(x,hist=False,kde=True,label="Normal")
    sns.distplot(y,hist=False,kde=True,label="Binomial")
    sns.distplot(z,hist=False,kde=True,label="Poisson")
    sns.distplot(uniform,hist=False,kde=True,label="Uniform")
    plt.show()
```

```python
compare()
```
